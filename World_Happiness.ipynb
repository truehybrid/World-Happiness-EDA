```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "world-happiness-report-2021.csv" 

df = pd.read_csv(data_path)

df.head()
```

```python
# Distribution of Happiness Scores
plt.figure(figsize=(8,5))
sns.histplot(df["Ladder score"], bins=20, kde=True)
plt.title("Distribution of Happiness Scores")
plt.xlabel("Happiness Score")
plt.ylabel("Count")
plt.savefig("happiness_distribution.png", bbox_inches="tight")
plt.show()

```

```python
# Top 10 Happiest Countries
top10 = df.nlargest(10, "Ladder score")
plt.figure(figsize=(10,6))
sns.barplot(data=top10, x="Ladder score", y="Country name", palette="viridis")
plt.title("Top 10 Happiest Countries (2021)")
plt.xlabel("Happiness Score")
plt.ylabel("")
plt.savefig("top10_happiness.png", bbox_inches="tight")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Happiness vs Factors)")
plt.savefig("correlation_heatmap.png", bbox_inches="tight")
plt.show()

```

```python
# Happiness vs GDP
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Logged GDP per capita", y="Ladder score")
plt.title("Happiness vs GDP per Capita")
plt.savefig("happiness_vs_gdp.png", bbox_inches="tight")
plt.show()

```

```python
# Happiness vs Social Support
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Social support", y="Ladder score")
plt.title("Happiness vs Social Support")
plt.savefig("happiness_vs_social.png", bbox_inches="tight")
plt.show()
```
